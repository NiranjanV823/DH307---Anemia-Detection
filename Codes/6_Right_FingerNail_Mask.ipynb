```python
import os
import cv2
```

```python
original_images_folder = "../Dataset Used/Right Fingernail_files/Images_right_fingernail"
```

```python
masks_folder = "../Dataset Used/Right Fingernail_files/right_nail_segmask/SegmentationClass"
```

```python
output_folder = "../Produced Images/Fingernail Masked"
```

```python
default_txt_path = "../Dataset Used/Right Fingernail_files/right_nail_segmask/ImageSets/Segmentation/default.txt"
```

```python
with open(default_txt_path, "r") as file:
    images_with_masks = file.read().splitlines()
```

```python
for image_name in images_with_masks:
  
    original_image_filename = image_name + ".jpg"  # Original images are .jpg
    mask_image_filename = image_name + ".png"     # Masks are .png
    
    original_image_path = os.path.join(original_images_folder, original_image_filename)
    mask_image_path = os.path.join(masks_folder, mask_image_filename)
    
   
    if not os.path.exists(original_image_path):
        print(f"Original image not found: {original_image_path}")
        continue
    if not os.path.exists(mask_image_path):
        print(f"Mask image not found: {mask_image_path}")
        continue
    
   
    original_image = cv2.imread(original_image_path)
    mask_image = cv2.imread(mask_image_path, cv2.IMREAD_GRAYSCALE)
    
    if original_image is None or mask_image is None:
        print(f"Error loading image or mask for {image_name}")
        continue
    
   
    binary_mask = mask_image < 125
    
  
    blue_channel, green_channel, red_channel = cv2.split(original_image)
    
    
    blue_channel[binary_mask] = 0
    green_channel[binary_mask] = 0
    red_channel[binary_mask] = 0
    
  
    masked_image = cv2.merge((blue_channel, green_channel, red_channel))
    
   
    output_image_path = os.path.join(output_folder, original_image_filename)
    cv2.imwrite(output_image_path, masked_image)
    print(f"Processed and saved: {output_image_path}")
```
