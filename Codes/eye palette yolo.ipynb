```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load the trained YOLO model
model = YOLO("../Trained Yolov5x\my_model.pt")

# Run inference on an image
results = model("data2.jpg")

# Extract and display the image with detections
for result in results:
    img = result.plot()  # Get image with bounding boxes

    # Convert from BGR (OpenCV format) to RGB (Matplotlib format)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Display the image
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis("off")
    plt.show()
```

```python
image_path = 'data2.jpg'
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis("off")
plt.show()
```
