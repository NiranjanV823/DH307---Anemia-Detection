```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the original image
original_image_path = "Original.jpg"
original_image = cv2.imread(original_image_path)

# Load the mask
mask_image_path = "mask.png"
mask_image = cv2.imread(mask_image_path, cv2.IMREAD_GRAYSCALE)  # Load as grayscale
```

```python
# Convert BGR to RGB for visualization
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

# Plot original image and mask
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(original_image_rgb)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Mask")
plt.imshow(mask_image, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Create a binary mask where pixels < 125 are considered part of the mask
binary_mask = mask_image < 125

# Split the original image into color channels
blue_channel, green_channel, red_channel = cv2.split(original_image)

# Apply the mask to each channel
blue_channel[binary_mask] = 0
green_channel[binary_mask] = 0
red_channel[binary_mask] = 0

# Merge the channels back into one image
masked_image = cv2.merge((blue_channel, green_channel, red_channel))
```

```python
# Convert BGR to RGB for visualization
masked_image_rgb = cv2.cvtColor(masked_image, cv2.COLOR_BGR2RGB)

# Plot the masked image
plt.figure(figsize=(5, 5))
plt.title("Masked Image")
plt.imshow(masked_image_rgb)
plt.axis("off")
plt.show()
```
