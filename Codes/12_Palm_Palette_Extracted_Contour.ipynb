```python
import cv2
import os
import numpy as np
```

```python
original_folder = "../Dataset Used/Left Palm_files/Images_left_palm"
```

```python
default_txt_path = "../Dataset Used/Left Palm_files/left_palm_segmasks/ImageSets/Segmentation/default.txt"
```

```python
output_folder = "../Produced Images/Palm Palette"
```

```python
os.makedirs(output_folder, exist_ok=True)
```

```python
with open(default_txt_path, "r") as file:
    image_names = file.read().splitlines()
```

```python
# Template Matching to detect the palette region
def detect_palette(image, template):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    result = cv2.matchTemplate(gray_image, template, cv2.TM_CCOEFF_NORMED)

    # Set a threshold for matching
    threshold = 0.8
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

    if max_val >= threshold:
        template_height, template_width = template.shape
        top_left = max_loc
        bottom_right = (top_left[0] + template_width, top_left[1] + template_height)
        return top_left, bottom_right
    else:
        return None
```

```python
# Load the template image (grayscale)
template_path = "Template Pallete.jpeg"  # Replace with the path to the palette template
template = cv2.imread(template_path, cv2.IMREAD_GRAYSCALE)
if template is None:
    raise FileNotFoundError(f"Template image not found: {template_path}")
```

```python
for image_name in image_names:
    input_image_path = os.path.join(original_folder, image_name + ".jpg")

    if not os.path.exists(input_image_path):
        print(f"Image not found: {input_image_path}")
        continue

    image = cv2.imread(input_image_path)
    if image is None:
        print(f"Failed to load image: {input_image_path}")
        continue

    # Detect the palette region using template matching
    palette_region = detect_palette(image, template)

    if palette_region:
        top_left, bottom_right = palette_region
        x1, y1 = top_left
        x2, y2 = bottom_right
        palette = image[y1:y2, x1:x2]

        output_image_path = os.path.join(output_folder, image_name + "_palette.jpg")
        cv2.imwrite(output_image_path, palette)
        print(f"Processed and saved: {output_image_path}")
    else:
        print(f"No palette detected for image: {input_image_path}")
```
