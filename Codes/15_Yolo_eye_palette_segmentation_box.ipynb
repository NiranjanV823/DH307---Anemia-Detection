```python
from ultralytics import YOLO
```

```python
zip_path = '/content/Images_left_eye.zip'
```

```python
extract_path = "/content/Images_left_eye"
```

```python
output_dir = "/content/output"
```

```python
default_file = "/content/default.txt"
```

```python
# Unzip the original images folder
if not os.path.exists(extract_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall("/content/")  # Extract to /content/
    print("✅ Images Unzipped Successfully!")
```

```python
os.makedirs(output_dir, exist_ok=True)
```

```python
model = YOLO('/content/my_model.pt')
```

```python
# Read image names from default.txt
with open(default_file, "r") as f:
    image_names = {line.strip() + ".jpg" for line in f.readlines()}  # Append .jpg
```

```python
# Process only images listed in default.txt
for image_name in image_names:
    image_path = os.path.join(extract_path, image_name)  # Use extracted folder

    if os.path.exists(image_path):  # Check if image exists
        # Perform inference using YOLO model
        results = model(image_path)

        # Save each processed image in the output folder
        for result in results:
            output_path = os.path.join(output_dir, image_name)
            result.save(filename=output_path)  # Save segmented image with same name
```
