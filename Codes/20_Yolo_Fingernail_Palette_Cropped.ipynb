```python
from ultralytics import YOLO
```

```python
dataset_path = '/kaggle/input/fingernail-dataset'
```

```python
original_images_path = os.path.join(dataset_path, "Images_right_fingernail/Images_right_fingernail")
```

```python
output_dir = "/kaggle/working/output"
```

```python
default_file = os.path.join(dataset_path, "default.txt")
```

```python
model_path = os.path.join(dataset_path, "my_model.pt")
```

```python
os.makedirs(output_dir, exist_ok=True)
```

```python
model = YOLO(model_path)
```

```python
with open(default_file, "r") as f:
    image_names = {line.strip() + ".jpg" for line in f.readlines()}  # Append .jpg
```

```python
for image_name in image_names:
    image_path = os.path.join(original_images_path, image_name)

    if os.path.exists(image_path):  # Check if image exists
        # Perform inference using YOLO
        results = model(image_path, imgsz=640)

        # Load the original image
        image = cv2.imread(image_path)

        # Process results and crop bounding boxes
        for result in results:
            for i, box in enumerate(result.boxes.xyxy):  # Loop through detected boxes
                x1, y1, x2, y2 = map(int, box)  # Convert to integers

                # Crop the detected object
                cropped_img = image[y1:y2, x1:x2]

                # Define output path (same filename as original image)
                output_path = os.path.join(output_dir, image_name)

                # Save cropped image
                cv2.imwrite(output_path, cropped_img)
```

```python
import shutil
shutil.make_archive("/kaggle/working/output", 'zip', output_dir)
```
