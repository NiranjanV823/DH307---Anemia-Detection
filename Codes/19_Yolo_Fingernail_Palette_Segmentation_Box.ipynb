```python
from ultralytics import YOLO
```

```python
dataset_path = '/kaggle/input/fingernail-dataset'
```

```python
original_images_path = os.path.join(dataset_path, "Images_right_fingernail/Images_right_fingernail")
```

```python
output_dir = "/kaggle/working/output"
```

```python
default_file = os.path.join(dataset_path, "default.txt")
```

```python
model_path = os.path.join(dataset_path, "my_model.pt")
```

```python
os.makedirs(output_dir, exist_ok=True)
```

```python
model = YOLO(model_path)
```

```python
with open(default_file, "r") as f:
    image_names = {line.strip() + ".jpg" for line in f.readlines()}  # Append .jpg
```

```python
# Process only images listed in default.txt
for image_name in image_names:
    image_path = os.path.join(original_images_path, image_name)  
    
    if os.path.exists(image_path):  # Check if image exists
        # Perform inference using YOLO model
        results = model(image_path)

        # Save each processed image in the output folder
        for result in results:
            output_path = os.path.join(output_dir, image_name)
            result.save(filename=output_path)  # Save segmented image with same name
```
